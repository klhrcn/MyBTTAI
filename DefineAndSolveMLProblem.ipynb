```python
#fill missing values
df=df.fillna(df.mean(numeric_only=True))

#drop columns
df=df.drop(columns=['country','year'],errors='ignore')

#set features
X=df.drop(columns=['Life Ladder'])
y=df['Life Ladder']


# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=1234)
nn_model=keras.Sequential([
    keras.layers.InputLayer(input_shape=X_train.shape[1]),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(32,activation='relu'),
    keras.layers.Dense(16,activation='relu'),
    keras.layers.Dense(1),
])

nn_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

class ProgBarLoggerNEpochs(keras.callbacks.Callback):
    def __init__(self, num_epochs: int, every_n: int = 10):
        self.num_epochs = num_epochs
        self.every_n = every_n
    def on_epoch_end(self, epoch, logs=None):
        if (epoch + 1) % self.every_n == 0:
            print(f"Epoch {epoch + 1}/{self.num_epochs} - Loss: {logs['loss']:.4f}, Val Loss: {logs['val_loss']:.4f}, MAE: {logs['mae']:.4f}")

num_epochs=55
history=nn_model.fit(
    X_train,y_train,
    epochs=num_epochs,
    validation_split=0.2,
    verbose=0,
    callbacks=[ProgBarLoggerNEpochs(num_epochs, every_n=5)]
)
                  
plt.plot(history.history['loss'], label='Training Loss')            
plt.plot(history.history['val_loss'],label='Valdiation Loss')
plt.xlabel=('Epoch')
plt.ylabel('MSE Loss')
plt.legend()
plt.show()
                  
                  
plt.plot(history.history['mae'], label='Training MAE')            
plt.plot(history.history['val_mae'],label='Valdiation MAE')
plt.xlabel=('Epoch')
plt.ylabel('MAE')
plt.legend()
plt.show()
```
